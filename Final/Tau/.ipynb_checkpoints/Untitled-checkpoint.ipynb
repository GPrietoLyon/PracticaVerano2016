```python
import matplotlib.pyplot as plt
import numpy as np
import math 
from astropy.io import ascii
```

```python
Mtot=1
e=math.e
```

```python
ages=ascii.read("age.txt") 
taus=ascii.read("tau.txt")
```

```python
tiempos=[]
for i in range(0,100):
    tiempos.append([ages[i][0],taus[i][0]])



```

```python
def sfh(t,tau):
    Y=(Mtot/(tau**2))*t*e**(-t/tau)
    return Y
```

```python
%matplotlib inline

for i in range(0,100):
    age=tiempos[i][0]
    tau=tiempos[i][1]
    t=np.linspace(0,age,1000)
    s=sfh(t,tau)
    plt.plot(t,s)
    plt.xlabel("tiempo",size=15)
    plt.ylabel("tau",size=15)
    plt.show()
    print "edad=",age,"  tau=",tau, "SFR max=",max(s)
    
    
```
