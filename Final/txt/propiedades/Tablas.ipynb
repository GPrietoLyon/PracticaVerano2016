```python
import numpy as np

def media(lista):
    a=[]
    b=[]
    c=[]
    for dato in lista:
        a.append(dato[0])
        b.append(dato[1])
        c.append(dato[2])
    return np.mean(a),np.mean(b),np.mean(c)
```
