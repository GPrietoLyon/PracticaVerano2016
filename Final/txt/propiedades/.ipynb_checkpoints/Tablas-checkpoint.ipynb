```python
from astropy.io import ascii
```

```python
archiv=["age.txt","mass.txt","Zmetal.txt","sfr.txt","av.txt"]


agefsps = ascii.read("fsps/"+archiv[0])  
massfsps = ascii.read("fsps/"+archiv[1])  
zmetalfsps = ascii.read("fsps/"+archiv[2])  
sfrfsps = ascii.read("fsps/"+archiv[3])  
avfsps = ascii.read("fsps/"+archiv[4])  

agebc03 = ascii.read("bc03/"+archiv[0])  
massbc03 = ascii.read("bc03/"+archiv[1])  
zmetalbc03 = ascii.read("bc03/"+archiv[2])  
sfrbc03 = ascii.read("bc03/"+archiv[3])  
avbc03 = ascii.read("bc03/"+archiv[4])  

agepega = ascii.read("pegase/"+archiv[0])  
masspega = ascii.read("pegase/"+archiv[1])  
zmetalpega = ascii.read("pegase/"+archiv[2])  
sfrpega = ascii.read("pegase/"+archiv[3])  
avpega = ascii.read("pegase/"+archiv[4])  

agemara = ascii.read("maraston/"+archiv[0])  
massmara = ascii.read("maraston/"+archiv[1])  
zmetalmara = ascii.read("maraston/"+archiv[2])  
sfrmara = ascii.read("maraston/"+archiv[3])  
avmara = ascii.read("maraston/"+archiv[4])


fsage,fsmass,fszme,fssfr,fsav=[],[],[],[],[]
bcage,bcmass,bczme,bcsfr,bcav=[],[],[],[],[]
peage,pemass,pezme,pesfr,peav=[],[],[],[],[]
maage,mamass,mazme,masfr,maav=[],[],[],[],[]
```

```python
def llenar(ascii,lista):
    for dato in ascii:
        lista.append([dato[0],  dato[1]-dato[0] , dato[0]-dato[2] ])
        
llenar(agefsps,fsage)
llenar(massfsps,fsmass)
llenar(zmetalfsps,fszme)
llenar(sfrfsps,fssfr)
llenar(avfsps,fsav)

llenar(agebc03,bcage)
llenar(massbc03,bcmass)
llenar(zmetalbc03,bczme)
llenar(sfrbc03,bcsfr)
llenar(avbc03,bcav)

llenar(agepega,peage)
llenar(masspega,pemass)
llenar(zmetalpega,pezme)
llenar(sfrpega,pesfr)
llenar(avpega,peav)

llenar(agemara,maage)
llenar(massmara,mamass)
llenar(zmetalmara,mazme)
llenar(sfrmara,masfr)
llenar(avmara,maav)
```

```python
Z = ascii.read("Z.txt") 
ID=[]
z=[]
for i in Z:
    ID.append(i[0])
    z.append(round(i[1],2))
```
