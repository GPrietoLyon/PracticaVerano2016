```python
from astropy.io import ascii
cat = ascii.read("cat.txt") 
c = ascii.read("highz.txt") 
```

```python
cand=[]
for i in c:
    cand.append(i[0])
```

```python
f105,f140,f160,f125,f814,z=[],[],[],[],[],[]
for i in range(0,len(cat)):
    f105.append(cat[i][12])
    f140.append(cat[i][8])
    f160.append(cat[i][6])
    f125.append(cat[i][10])
    f814.append(cat[i][14])
    z.append(cat[i][3])
```

```python
#z = 7-8
falsos78=0
totalno78=0

for i in range(0,len(cat)):
    
    

    
    a=0
    b=0
    c=0


    if z[i]<7 or z[i]>8:
        totalno78=totalno78+1
        
        print "ID: ",cand[i]
        print "z",z[i]
        
    
        if f814[i]-f105[i]>0.8:
            print "check 1"
            a=1
        if f105[i]-f125[i]<0.6:
            print "check 2"
            b=1
        if f814[i]-f105[i]>0.8+(f105[i]-f125[i]):
            print "check 3"
            c=1
        if a==1 and b==1 and c==1:
            falsos78=falsos78+1
            print "ALL CHECK"
        
        print "\n \n"
        
    
    
```

```python
#z = 7-8
correctos78=0
total78=0

for i in range(0,len(cat)):
    
    

    
    a=0
    b=0
    c=0


    if z[i]>7 and z[i]<8:
        total78=total78+1
        
        print "ID: ",cand[i]
        print "z",z[i]
        
    
        if f814[i]-f105[i]>0.8:
            print "check 1"
            a=1
        if f105[i]-f125[i]<0.6:
            print "check 2"
            b=1
        if f814[i]-f105[i]>0.8+(f105[i]-f125[i]):
            print "check 3"
            c=1
        if a==1 and b==1 and c==1:
            correctos78=correctos78+1
            print "ALL CHECK"
        
        print "\n \n"
        
```

```python
print "Total con Z=7-8: ", total78
print "Encontrados :", correctos78
print "Total no Z=7-8",totalno78
print "Falsos positivos",falsos78
```
