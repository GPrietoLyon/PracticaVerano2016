```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

# T-model and delayed T-model

```python
e=math.e

def tmod(t,tau):
    Ys=(1.0/tau)*(e**(t/tau))
    return Ys

def deltmod(t,tau):
    Ys=(1.0/((tau)**2.0))*t*(e**(t/tau))
    return Ys

```

```python
t=np.linspace(0,10,1000)
tau=1.0
y1=deltmod(t,tau)
plt.plot(t,y1)
plt.show()
print y1
```
